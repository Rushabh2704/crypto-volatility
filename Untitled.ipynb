```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
```

```python
df = pd.read_parquet('data/processed/features_labeled.parquet')
print(f"Shape: {df.shape}")
df.head()
```

```python
df.describe()
```

```python
print(df.shape)
print(df.dtypes)
print(df["label"].value_counts())
print(df.isnull().sum())
```

```python
df['label'].value_counts().plot(kind='bar', color=['steelblue', 'red'])
plt.title('Label Distribution (0=No Spike, 1=Spike)')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('reports/eda_label_distribution.png')
plt.show()
```

```python
vol = df['sigma_future'].dropna()
percentiles = range(50, 100)
values = [np.percentile(vol, p) for p in percentiles]

plt.figure(figsize=(10, 5))
plt.plot(list(percentiles), values)
plt.title('Future Volatility Percentiles')
plt.xlabel('Percentile')
plt.ylabel('σ_future')
plt.axvline(x=90, color='red', linestyle='--', label='τ = 0.000026 (90th percentile)')
plt.legend()
plt.tight_layout()
plt.savefig('reports/eda_volatility_percentiles.png')
plt.show()
print(f"90th percentile: {np.percentile(vol, 90):.6f}")
print(f"95th percentile: {np.percentile(vol, 95):.6f}")
```

```python
features = ['spread', 'book_imbalance', 'rolling_volatility', 'volume_change']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for ax, feat in zip(axes.flatten(), features):
    df[feat].hist(bins=50, ax=ax)
    ax.set_title(feat)

plt.tight_layout()
plt.savefig('reports/eda_feature_distributions.png')
plt.show()
```

```python
cols = ['spread', 'book_imbalance', 'rolling_volatility', 'volume_change', 'midprice_return', 'label']
corr = df[cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.savefig('reports/eda_correlation_heatmap.png')
plt.show()
```

```python
plt.figure(figsize=(14, 4))
plt.plot(df['rolling_volatility'].values)
plt.title('Rolling Volatility Over Time')
plt.xlabel('Tick')
plt.ylabel('Volatility')
plt.tight_layout()
plt.savefig('reports/eda_volatility_over_time.png')
plt.show()
```
